```python
dropcolumn = list(df.filter(regex='footnote'))
df_clean = df[df.columns.drop(dropcolumn)]
df_clean.head()
```

```python
new_df= df_clean.replace('Not Available', np.nan )
new_df
```

```python
#converting to numerial values
ordinal_col = list(new_df.filter(regex="national comparison"))

mapper = {
    'Below the national average' : 0,
    'Same as the national average' : 1, 
    'Above the national average' : 2
}
for col in ordinal_col:
    new_df.loc[:, new_df.columns == col]= new_df.loc[:, new_df.columns == col].replace(mapper)
new_df.head() 
```

```python
DF = new_df.dropna()
```

```python
noduplicate = DF.drop_duplicates(['Provider ID','ZIP Code'])
utility_value = DF.pivot_table(values='Hospital overall rating', index='ZIP Code', columns='Provider ID', aggfunc='first',fill_value=0)
utility_value.head()
```

```python
utility_value.shape
```

```python
X1 = utility_value.T
X1.head()
```

```python
X1.shape
```

```python
X=X1
```

```python
#converting to truncated SVD
import sklearn
from sklearn.decomposition import TruncatedSVD

SVD = TruncatedSVD(n_components=10)

# model fitting
decomposed_matrix = SVD.fit_transform(X1)
decomposed_matrix.shape
```

```python
correlation_matrix = np.corrcoef(decomposed_matrix)
correlation_matrix.shape
```

```python
classifiers = {
    "Support Vector Classifier": SVC()
}
```

```python
#Determining accuracy
from sklearn.model_selection import cross_val_score
for key, classifier in classifiers.items():
    classifier.fit(X , y)
    acc_score = cross_val_score(classifier, X, y, cv=5)
    print("Classifiers: ", classifier.__class__.__name__, "Has a accuracy score of", round(acc_score.mean(), 2), "accuracy score")
    print("Classifiers: ", classifier.__class__.__name__, "Has a accuracy score percentage of", round(acc_score.mean(), 2) * 100, " % accuracy percentage")
```
